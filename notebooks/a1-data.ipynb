```python
releases = sell.groupby(['store_id','item_id'])['wm_yr_wk'].agg(['min']).reset_index()
releases.columns = ['store_id','item_id','release']
releases
```

```python
melted = (
    melted.merge(releases, how="left", on=['store_id','item_id'])
     .merge(calendar.df[["d", "wm_yr_wk"]], how="left", on=["d"])
    
)
```

```python
melted = (
    melted.query("wm_yr_wk >= release").reset_index(drop=True)
    .assign(release=lambda df: (df.release - df.release.min()).astype(np.int16))
)
melted.to_parquet(fintopmet.fp.DATA / "m5" / "proc" /"grid_part1.parquet")
```

```python
sell = sell.assign(
    price_max = sell.groupby(['store_id','item_id'])['sell_price'].transform('max'),
    price_min = sell.groupby(['store_id','item_id'])['sell_price'].transform('min'),
    price_std = sell.groupby(['store_id','item_id'])['sell_price'].transform('std'),
    price_mean = sell.groupby(['store_id','item_id'])['sell_price'].transform('mean'),
    price_norm = lambda x: x['sell_price'] / x.groupby(['store_id','item_id'])['sell_price'].transform('max'),
    price_nunique = sell.groupby(['store_id','item_id'])['sell_price'].transform('nunique'),
    item_nunique = sell.groupby(['store_id','sell_price'])['item_id'].transform('nunique')
).merge(
    calendar.df[['wm_yr_wk','month','year']].drop_duplicates(subset=['wm_yr_wk'])[['wm_yr_wk','month','year']],
    on=['wm_yr_wk'],
    how='left'
).assign(
    price_momentum = lambda df: df['sell_price'] / df.groupby(['store_id','item_id'])['sell_price'].transform(lambda df: df.shift(1)),
    price_momentum_m = lambda df: df['sell_price'] / df.groupby(['store_id','item_id','month'])['sell_price'].transform('mean'),
    price_momentum_y = lambda df: df['sell_price'] / df.groupby(['store_id','item_id','year'])['sell_price'].transform('mean')
)
```

```python
melted = melted.merge(sell, on=['store_id','item_id','wm_yr_wk'], how='left')
melted.to_parquet(fintopmet.fp.DATA / "m5" / "proc" /"grid_part2.parquet")
```

```python
event_cols = [
         'event_name_1',
         'event_type_1',
         'event_name_2',
         'event_type_2',
         'snap_CA',
         'snap_TX',
         'snap_WI'
]
icols = ['date', 'd'] + event_cols

melted = (
    melted.merge(calendar.df[icols], on=['d'], how='left')
    .astype({col: "category" for col in event_cols})
)
```

```python
import numpy as np
from math import ceil

melted = melted.assign(
    date = lambda df: pd.to_datetime(df['date']),
    tm_d = lambda df: df['date'].dt.day.astype(np.int8),
    tm_w = lambda df: df['date'].dt.isocalendar().week.astype(np.int8),
    tm_m = lambda df: df['date'].dt.month.astype(np.int8),
    tm_y = lambda df: (df['date'].dt.year - df['date'].dt.year.min()).astype(np.int8),
    tm_wm = lambda df: df['tm_d'].apply(lambda x: ceil(x/7)).astype(np.int8),
    tm_dw = lambda df: df['date'].dt.dayofweek.astype(np.int8),
    tm_w_end = lambda df: (df['date'].dt.dayofweek >= 5).astype(np.int8)
)
```
